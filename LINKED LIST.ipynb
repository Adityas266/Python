```python
class Node:
    def __init__(self, data=0, next=None):
        self.data = data
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None
        
    def insertLeft(self, value):
        new = Node(value)
        if not self.head:
            self.head = new
        else:
            new.next = self.head
            self.head = new
    
    def deleteLeft(self):
        if not self.head:
            return -1
        node = self.head
        val = node.data
        self.head = self.head.next
        del node
        return val
    
    def insertRight(self, value):
        new = Node(value)
        if not self.head:
            self.head = new
        else:
            cur = self.head
            while cur.next:
                cur = cur.next
            cur.next = new
    
    def deleteRight(self):
        if not self.head:
            return -1
        
        if not self.head.next:
            tmp = self.head.data
            self.head = None
            return tmp
        
        cur = self.head
        while cur.next.next:
            cur = cur.next
        node = cur.next
        val = node.data
        cur.next = None
        del node
        return val
    
    def searchIndex(self, x):
        cur = self.head
        index = 0
        while cur:
            if cur.data==x:
                return index
            cur = cur.next
            index += 1
        return -1
    
    def search(self, x):
        cur = self.head
        while cur:
            if cur.data==x:
                return True
            cur = cur.next
        return False
    
    def deleteKey(self, x):
        if not self.head:
            return False
        if self.head.data==x:
            self.head = self.head.next
            return True
        p,q = self.head,self.head.next
        while q and q.data!=x:
            p = q
            q = q.next
        if q:
            p.next = q.next
            return True
        return False
    
    
    def printList(self):
        cur = self.head
        while cur:
            print(f'| {cur.data} |', end=' => ')
            cur = cur.next
        print(None, end='\n\n')


ll = LinkedList()
print("Linked List Created!!")

while True:
    print("\n1. Insert Left\n2. Insert Right\n3. Delete Left\n4. Delete Right\n5. Delete Key\n6. Search\n7. Search Index\n8. Print Linked List")
    ip = input("Enter your Choice: ")
    
    if ip=='1':
        ll.insertLeft(int(input("Enter the value: ")))
        print("Successfully inserted...")
    
    elif ip=='2':
        ll.insertRight(int(input("Enter the value: ")))
        print("Successfully inserted...")
    
    elif ip=='3':
        val = ll.deleteLeft()
        if val==-1:
            print("Linked List Empty!!")
        else:
            print(f"Deleted Node: {val.data}")
    
    elif ip=='4':
        val = ll.deleteRight()
        if val==-1:
            print("Linked List Empty!!")
        else:
            print(f"Deleted Node: {val.data}")
    
    elif ip=='5':
        val = ll.deleteKey(int(input("Enter the key: ")))
        if not val:
            print("Key not found in the Linked List!!")
        else:
            print(f"Key:{val} Deleted!")
    
    elif ip=='6':
        val = ll.search(int(input("Enter the value: ")))
        if not val:
            print("Element not found!")
        else:
            print("Element found!!")
    
    elif ip=='7':
        ind = ll.searchIndex(int(input("Enter the value: ")))
        if ind==-1:
            print("Element not found")
        else:
            print(f"Element found at index: {ind}")
    
    elif ip=='8':
        print("Nodes in the Linked List: ")
        ll.printList()
    
    else:
        print("Your Final Linked List: ")
        ll.printList()
        break
```
