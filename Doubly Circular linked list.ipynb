```python
class Node:
    def _init_(self,val):
        self.val = val
        self.right = None
        self.left = None
class CircularDoublyLinkedList:
    def _init_(self):
        self.head = self.last = None
    def insertLeft(self,val):
        n = Node(val)
        if self.head == None:
            self.head = self.last = n
        else:
            n.right = self.head
            self.head.left = n
            self.head = n
        self.last.right = self.head
        self.head.left = self.last
    def insertRight(self,val):
        n = Node(val)
        if self.head == None:
            self.head = self.last = n
        else:
            self.last.right = n
            n.left = self.last
            self.last = n
        self.last.right = self.head
        self.head.left = self.last
    def delLeft(self):
        if self.head==None:
            return -1
        else:
            t = self.head
            self.head = self.head.right
            self.last.right = self.head
            self.head.left = self.last
            return t.val
    def delRight(self):
        if self.head==None:
            return -1
        else:
            temp = self.head
            if self.head==self.last:
                self.head = self.last = None
            else:
                while temp!=self.last:
                    temp = temp.right
                if temp == self.head:
                    self.head = self.last = None
                self.last = temp.left
                self.last.right = self.head
                self.head.left = self.last
            return temp.val
    def printList(self):
        if self.head==None:
            return -1
        temp = self.head
        #while True:
        while temp:
            print('|',temp.val,'|->',end=" ")
            if temp==self.last:
                break
            temp = temp.right
        print()
    def printReverse(self):
        if self.head==None:
            return -1
        t = self.last
        while t:
            print('|',t.val,'|->',end=" ")
            if t == self.head:
                break
            t = t.left
        print()

l = CircularDoublyLinkedList()
while True:
    print("1.InsertLeft\n2.InsertRight\n3.deleteLeft\n4.deleteRight\n5.printList\n6.printReverse\n0.Exit")
    n = input("Enter your choice: ")
    if n=='1':
        val = input("Enter Data:")
        l.insertLeft(val)
        print("Element Inserted at Left:",val)
    elif n=='2':
        val = input("Enter Data:")
        l.insertRight(val)
        print("Element Inserted at Right:",val)
    elif n=='3':
        t = l.delLeft()
        if t!=-1:
            print("Element Deleted:",t)
        else:
            print("LinkedList is Empty")
    elif n=='4':
        t = l.delRight()
        if t!=-1:
            print("Element Deleted:",t)
        else:
            print("LinkedList is Empty")
    elif n=='5':
        t = l.printList()
        if t==-1:
            print("LinkedList is Empty")
    elif n=='6':
        t = l.printReverse()
        if t==-1:
            print("LinkedList is Empty")
    elif n=='0':
        print("Thank you")
        break
    else:
        print("Invalid..")
```
